#### 연습문제 1
- 타이타닉 호 승객에 대해서, 성별(sex) 인원수, 나이별(age) 인원수, 선실별(class) 인원수, 사망/생존(alive)인원수를 구하시오.

- 성별 인원수는 인덱스 기준으로 정렬하고
- 나이별 인원수는 값 기준 정렬
- 그 나머지는 임의 기준 정렬 하시오.

```python
# 성별 인원수는 인덱스 기준으로 정렬 (오름차순 정렬)
titanic['sex'].value_counts().sort_index()
```

```python
# 나이별 인원수 값 기준 정렬 / 나이 = 연속형, 실수형 데이터
titanic['age'].value_counts()

# 빈도수가 많은 나이부터 내림차순
titanic['age'].value_counts().sort_values(ascending=False)

# dot(.) 연산자
titanic.age.value_counts().sort_values(ascending=False)
```

```python
# 선실별(class) 인원수 index 기준 내림차순 정렬
titanic['class']

# titanic.class 
# - SyntaxError: invalid syntax 문법오류, 예약어를 일반 변수처럼 사용했을 경우
# class 컬럼명 사용은 가능
# 단, class 단어가 예약어 - .class 형태로 사용하면 컬럼명이 아닌 예약어로 확인됨

titanic['class'].value_counts().sort_index(ascending=False)
```

```python
# 사망/생존 인원 구하기
titanic.alive.value_counts().sort_values()
```

```python
# 튜플 생성하기 (,를 기준으로 튜플 생성됨)
cols = ('sex', ); type(cols)
cols = 'sex', ; type(cols)

# 튜플 생성 시 소괄호 생략 가능
cols = ('sex', 'age', 'class', 'alive')
cols = 'sex', 'age', 'class', 'alive'
type(cols)

v_cnt = {}
for col in cols:
    v_cnt[col] = titanic[col].value_counts()
v_cnt

# value_counts key로 확인
v_cnt['sex']
```

#### 연습문제 3
- 타이타닉호 승객을 사망자와 생존자 그룹으로 나누고(alive)
- 각 그룹에 대해 미성년자, 청년, 중년, 장년, 노년 승객의 비율을 구하시오.
    - bins=[1,15,25,35,60,99]
    - labels=['미성년자','청년','중년','장년','노년']
    - 각 그룹별 비유의 전체 합은 1이 되어야 한다.

```python
# 사망자 생존자 data 분리
no = titanic[titanic.alive=='no']
yes = titanic[titanic.alive=='yes']
no.head()
yes.head()
```

```python
labels = ['미성년자', '청년', '중년', '장년', '노년']
bins = [1, 15, 25, 35, 60, 99]

# bins 유효성 검사 - 최소값, 최대값 검사
np.min(titanic.age)  # 0.42 -> 최소범위 1로 시작하면 모든 나이 포함 X (유효성 오류)
np.max(titanic.age)  # 80.0 -> 최대범위 99이므로 포함 O

# bins 수정
bins = [0, 15, 25, 35, 60, 99]
```

```python
# 카테고리 데이터 생성 - 각 df마다 생성 (no, yes)
no['age_cut'] = pd.cut(no.age, bins, labels=labels)
yes['age_cut'] = pd.cut(yes['age'], bins, labels=labels)

no.head()  # 결측치 먼저 제거 후 실행하기
yes.head()
```

```python
# 각 그룹에 대해 미성년자, 청년, 중년, 장년, 노년 승객의 비율을 구하시오
no['age_cut'].value_counts(normalize=True)
yes['age_cut'].value_counts(normalize=True)

no['age_cut'].value_counts(normalize=True).sum()
yes['age_cut'].value_counts(normalize=True).sum()
```
