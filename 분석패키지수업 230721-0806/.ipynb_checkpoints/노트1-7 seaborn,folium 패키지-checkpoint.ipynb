#### seaborn 스타일링 

- set_style: background color, grid, spine, tick을 정의하여 그림의 전반적인 모양을 스타일링
- set_context: 프리젠테이션이나 보고서와 같은 다양한 매체에 활용할 수 있도록 스타일링


1. set_style: 그림의 전반적인 모양 스타일링
    - Built-in Themes (내장 테마) 활용하기

        - Seaborn에는 5가지 기본 제공 테마가 있음 
        - darkgrid, whitegrid, dark, white, ticks. 
        - 기본값은 darkgrid이지만, 원하는대로 변경이 가능
2. sns.despin() 함수
    - 축선 표시 여부 결정하는 함수(right=T/F,left=T/F,top=T/F,bottom=T/F)
        - 각 속성값이 
        - True 면 : 표시하지 않는다.
        - False 면: 표시함
3. sns.set_context() 함수    
    - 전체 스케일 조정
    - paper, notebook, talk, poster. 여기서 기본값은 notebook이다.
    - 스케일로도 전반적인 사이즈를 조정할 수 있지만 실제 폰트 크기를 조정하는 파라미터는 별도로 있음

    - sns.set_context() 안에 font_scale 사용

```python
# seaborn이용한 style 설정

x=np.linspace(1,14,100)
y1=np.sin(x)
y2=2*np.sin(x)
y3=3*np.sin(x)
y4=4*np.sin(x)

sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
plt.plot(x,y1,x,y2,x,y3,x,y4)
plt.show()
```

```python
# seaborn이용한 style 설정

x=np.linspace(1,14,100)
y1=np.sin(x)
y2=2*np.sin(x)
y3=3*np.sin(x)
y4=4*np.sin(x)

sns.set_style('darkgrid')  # 배경 회색
plt.figure(figsize=(10,6))
plt.plot(x,y1,x,y2,x,y3,x,y4)
plt.show()
```

```python
# sns.despin() 함수
# 축선 표시 여부 결정하는 함수(right=T/F,left=T/F,top=T/F,bottom=T/F)
# 각 속성값이 True 면 : 표시하지 않는다.
#           False 면: 표시함

x=np.linspace(1,14,100)
y1=np.sin(x)
y2=2*np.sin(x)
y3=3*np.sin(x)
y4=4*np.sin(x)


sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
plt.plot(x,y1,x,y2,x,y3,x,y4)

# sns.despine(right=False, left=False)  # top = True, bottom = False
# sns.despine(right=True, left=True, top=True, bottom=True)
sns.despine(right=True, left=False, top=True, bottom=False)
plt.show()
```

```python
# seaborn이용한 style 설정

x=np.linspace(1,14,100)
y1=np.sin(x)
y2=2*np.sin(x)
y3=3*np.sin(x)
y4=4*np.sin(x)

# set_context(paper/ notebook/talk/ poster)
sns.set_context('talk')
plt.figure(figsize=(10,6))
plt.plot(x,y1,x,y2,x,y3,x,y4)
plt.show()
```

```python
# seaborn이용한 style 설정

x=np.linspace(1,14,100)
y1=np.sin(x)
y2=2*np.sin(x)
y3=3*np.sin(x)
y4=4*np.sin(x)

# paper, notebook, talk, poster.
plt.figure(figsize=(10,6))
sns.set_context('talk', font_scale=.5)
plt.plot(x,y1,x,y2,x,y3,x,y4)
plt.show()
```

###  Distribution Plot
- Distribution Plot은 데이터의 분포를 시각화하는데 도움이 됨
- 이 그래프를 사용하여 데이터의 평균(mean), 중위수(median), 범위(range), 분산(variance), 편차(deviation) 등을 이해 할 수 있음

1. Hist Plot
- 변수에 대한 히스토그램을 표시한다.
- 하나 혹은 두 개의 변수 분포를 나타내는 전형적인 시각화 도구로 범위에 포함화는 관측수를 세어 표시한다.

```python
sns.histplot(x=tips['total_bill'])
```

2. KDE Plot
- 하나 혹은 두 개의 변수에 대한 분포를 그림
- histplot은 절대량이라면 kdeplot은 밀도 추정치를 시각화
    - 결과물로는 연속된 곡선의 그래프를 얻을 수 있음

```python
sns.kdeplot(x=tips['total_bill'])
```

3. Box Plot
- 최대(maximum), 최소(minimum), mean(평균), 1 사분위수(first quartile), 3 사분위수(third quartile)를 보기 위한 그래프
- 특이치(outliar)를 발견하기에도 좋다.
- 단일 연속형 변수에 대해 수치를 표시하거나, 연속형 변수를 기반으로 서로 다른 범주형 변수를 분석할 수 있다.

```python
# tips 데이터를 이용한 box plot 그리기
# seaborn 패키지의 boxplot() 이용해서 그리기

sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
sns.boxplot(x=tips['total_bill'])
sns.despine()
plt.show()
```

```python
# tips 데이터를 이용한 box plot 그리기
# seaborn 패키지의 boxplot() 이용해서 그리기
# 요일별 영수증 금액의 분포 확인

sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
sns.boxplot(x=tips['day'], y=tips['total_bill'])
sns.despine(offset=10)  # x,y 축간의 간격을 설정하는 인수
plt.show()
```

```python
# tips 데이터를 이용한 box plot 그리기
# seaborn 패키지의 boxplot() 이용해서 그리기
# 요일별 영수증 금액의 분포 확인
# palette 인수 사용 예제 : palette = '파레트명'

sns.set_style('whitegrid')

plt.figure(figsize=(10,6))
# sns.boxplot(x=tips['day'],y=tips['total_bill'],palette='Set2')
sns.boxplot(x='day', y='total_bill', data=tips, palette='Set2')
plt.show()
```

```python
# hue 인수 : 카테고리값을 가지는 변수의 이름 지정하면
# 카테고리에 따라 그룹을 만들어서 표시

plt.figure(figsize=(10,6))
sns.boxplot(x='day', y='tip', hue='smoker', data=tips, palette='Set3')
plt.show()
```

```python
# hue 인수 : 카테고리값을 가지는 변수의 이름 지정하면
# 카테고리에 따라 그룹을 만들어서 표시
# y값을 tip 필드로 변경 후 비교

plt.figure(figsize=(10,6))
sns.boxplot(x='day',y='total_bill', hue='sex',data=tips, palette='Set3')
plt.show()
```

```python
# 요일별 팁의 분포를 비교하는 box plot을 그리시오. 파레트 Set3
sns.boxplot(x='day', y='tip', data=tips, palette='Set3')
```

```python
# 요일별 팁의 분포를 비교하되 성별에 따라 분포를 
# 나누어 표시하시오. 팔레트 세트 3

sns.boxplot(x='day', y='tip', hue='sex', data=tips, palette='Set3')
```

# seaborn.lmplot() 
- 스캐터그래프를 그릴때 회귀선을 기본으로 출력해주는 그래프 함수
- size 인수 : 그래프 크기
    - height 인수로 변경해서 사용하라는 메시지가 출력될 수 있음  
- hue 인수 : 카테고리 데이터를 분류하여 그룹화하되 하나의 그래프에 두개의 값을 모두 표현
- col 인수 : 카테고리 데이터를 분류하여 그룹화하되 그룹의 개수만큼 그래프를 생성해서 표현한다.


```python
# dataset : tips 사용
sns.set_style('darkgrid')
# plt.scatter(x=tips['total_bill'], y=tips['tip'])
sns.lmplot(x='total_bill', y='tip', data=tips, height=4)
plt.show()  # 양의 상관관계가 보인다.
```

```python
# dataset : tips 사용
# 회귀선 생략 코드 : 인수 :  fit_reg=False
sns.set_style('darkgrid')
sns.lmplot(x='total_bill', y='tip',fit_reg=False,
           data=tips, height=4)
plt.show()
```

```python
# hue는 그룹을 만듬
sns.lmplot(x='total_bill', y='tip', fit_reg=True, hue='smoker',
           data=tips)
plt.show()

# no의 상관정도가 더 높게 나타난다.
```

```python
# 영수증 금액과 팁의 관계성을 lmplot()으로 표현하되, 
# 성별을 구분해서 표현하시오.
sns.set_style('darkgrid')

sns.lmplot(x='total_bill', y='tip', fit_reg=True, hue='sex',
           data=tips)
plt.show()
```

```python
# 영수증 금액과 팁의 관계성을 lmplot()으로 표현하되, 
# 성별을 구분해서 표현하되 그래프를 각각 그리시오.
# hue 인수 : 카테고리 데이터를 분류하여 그룹화하되 하나의 
# 그래프에 두개의 값을 모두 표현
# col 인수 : 카테고리 데이터를 분류하여 그룹화하되 그룹의  
# 개수만큼 그래프를 생성해서 표현한다.

sns.set_style('darkgrid')
sns.lmplot(x='total_bill', y='tip', fit_reg=True, col='sex',  # col 그래프 분리
           data=tips)

plt.show()
```

```python
# 영수증 금액과 팁의 관계성을 lmplot()으로 표현하되, 
# 성별을 구분해서 표현하되 그래프를 각각 그리시오.
# hue 인수 : 카테고리 데이터를 분류하여 그룹화하되 하나의 
# 그래프에 두개의 값을 모두 표현
# col 인수 : 카테고리 데이터를 분류하여 그룹화하되 그룹의  
# 개수만큼 그래프를 생성해서 표현한다.

sns.set_style('darkgrid')
sns.lmplot(x='total_bill', y='tip',fit_reg=True, col='smoker',
           data=tips)
plt.show()
```

```python
# sns 패키지의 filghts data set 을 이용해서 heatmap 작성
flights=sns.load_dataset('flights')
flights.head()
flights.info()  # 144 entries, 144개월 (10년 이상의 데이터)
```

```python
# pivot 테이블을 이용해서 연도별 각 월 승객수로 구조를 변환
flight_p = pd.pivot_table(flights,
                         index='month',
                         columns='year',
                         fill_value=0)
# pivot table
flight_p
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(flight_p)
plt.title('year/month')
plt.show()

# 오른쪽으로 갈수록 점점 밝아짐
# 연도가 지날수록 승객수는 많아진다.
# 각 연도에 해당하는 열방향 데이터는 1월, 12월은 7, 8월에 비해 더 진한 색으로 나타난다.
# 즉, 한 해 동안 7, 8월에 이용 승객 수가 가장 많다.
```

```python
# 히트맵 그래프 위에 실제 데이터출력
# 사용인수 : annot=True, fmt=d/f - 출력데이터 형태

plt.figure(figsize=(10,6))
sns.heatmap(flight_p, annot=True, fmt='d')
plt.title('year/month')
plt.show()
```

2. 마커 생성 후 부착 하는 코드
    - folium.Marker().add_to(지도객체변수)

# 단계 구분도
- 데이터를 지도 그림에 반영시켜서 전달하는 그래프
- 보통 folium 에 layer로 얹게 된다.
- map 객체의 choropleth() 이용해서 작업
- 사용 인수
    - geo_data : 지도 파일 경로와 파일명
    - data : 지도에 표현되어야 할 값 변수
    - columns =  [key로 사용할 data, 실제 data의 필드명]
    - key_on : 지도 경계파일인 json에서 사용할 키 값이며 지도 Data는 표준 형식으로 만들어 져야 함
        - key_on 지칭 문법 : feature(키워드).json에서 나타나키 필드명
- folium.LayerControl().add_to(map)  : 단계구분도를 map에 표시하는 함수



```python
# 미국 각 주별 실업율을 data로 단계 구분도 작성
# 지도 경계 data가 있어야 함.json 파일로 작성됨
```

```python
# 관련 data 가져오기
state_unemploy = './data/data/02. folium_US_Unemployment_Oct2012.csv'
state_unemploy_data = pd.read_csv(state_unemploy)
state_unemploy_data.head()

# 50개주의 실업율 data
```

```python
# 지도 데이터
state_geo = './data/data/02. folium_us-states.json'
state_geo
```

```python
map = folium.Map(location=[40, -98], zoom_start=3)
map

# map 위에 단계 구분도 올리기
map.choropleth(geo_data=state_geo,
              data=state_unemploy_data,
              columns=['State', 'Unemployment'],
              key_on = 'feature.id',
              fill_color='YlGn',
              legend_name='Unemployment Rate(%)')

folium.LayerControl().add_to(map)
map
```

```python
# 한국 지도 데이터 이용해서 단계 구분도 그리기
import json

# 한국지도 경계의 json data 파일 가져오기
# 한글이 포함된경우에는 그냥 path만 연결하면 인코딩 문제 발생
# json.load( encoding='utf-8') 이용해서 변수에 data 저장 후 사용하는 것이 일반적

geo_path = './data/data/02. skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
# geo_str

# 서울을 중심으로 기본 지도 출력
# 위/경도[37.5502,126.982]
map = folium.Map(location=[37.5502,126.982], 
                 zoom_start=11,
                tiles='Stamen Toner')
map

# 단계 구분도 data 읽어오기
# csv 읽어오기 : pd.read_csv(경로포함파일명,encoding=,index_col='지정인덱스 필드명')
f_test = pd.read_csv('./data/data/cctv_result.csv', 
                     encoding='utf-8', index_col='구별')
f_test.head()
```
