```python
# cctv data 읽어오기
cctv_seoul = pd.read_csv('./data/data/01. CCTV_in_Seoul.csv')
cctv_seoul.head()
```

```python
# 인구데이터 가져오기 - 엑셀파일
# pd.read_excel('파일명')

# import_optional_dependency("xlrd", extra=err_msg) 에러 발생
!pip install xlrd

pop_Seoul = pd.read_excel('./data/data/01.population_in_Seoul.xls')
pop_Seoul.head()
```

```python
# 엑셀은 행, 열 원하는 부분만 가지고 올 수 있음
# 필요 data만 추출
# 사용할 열 : 자치구, 전체 인구수, 한국인, 외국인, 고령자(B, D, G, J, N)
# usecols = '선택열'
# header = 2 행을 header로 처리

pop_Seoul = pd.read_excel('./data/data/01.population_in_Seoul.xls',
                         header = 2,
                         usecols = 'B, D, G, J, N')
pop_Seoul.head()
```

```python
# 컬럼명 변경
pop_Seoul.rename(columns={pop_Seoul.columns[0]:'구별',
                          pop_Seoul.columns[1]:'인구수',
                          pop_Seoul.columns[2]:'한국인',
                          pop_Seoul.columns[3]:'외국인',
                          pop_Seoul.columns[4]:'고령자'}, 
                inplace=True)
pop_Seoul.head()
```

```python
# 각 데이터 파악하기
cctv_seoul.info()
pop_Seoul.info()

# 인구데이터는 불필요한 행 2개가 있음
```

```python
# 인구데이터 0번행은 합계 데이터 -> 삭제
pop_Seoul.drop([0], inplace=True)
pop_Seoul.head()
```

```python
# 구별 데이터가 중복되지 않았는지 확인
pop_Seoul.구별.unique()  # 중복 데이터 제외 후 반환
len(pop_Seoul.구별.unique())

# nan이 포함된 행이 있다.
```

```python
# 결측치 있는 행 제거 시 isnull() 활용
# isnull() : null/True 반환
pop_Seoul[pop_Seoul.구별.isnull()]
```

```python
# 26행 제거
pop_Seoul.drop(26, inplace=True)
pop_Seoul.info()
```

```python
pop_Seoul.구별.unique()
set(cctv_seoul.구별)

# 25개구 data만 존재
```

### 가공필드 생성, 절대 data를 이용해서 몇 가지 내용 확인

```python
cctv_seoul.head(2)
```

```python
# 설치대수가 적은 5개 구 확인
cctv_seoul.sort_values(by='소계').head()

cctv_seoul.sort_values(by='소계', ascending=False).head()

# => 도봉구의 설치대수가 가장 적다.
```

```python
# 설치대수가 많은 5개 구 확인
cctv_seoul.sort_values(by='소계', ascending=False).head(10)

# => 강남구의 설치대수가 가장 많다.
```

### 최근 증가율 가공필드 생성
- 설치대수 절대량이 도봉구가 가장 적다
    - 도봉구는 설치를 늘리지 않고 있는지 확인
- 최근 증가율 = (3개년 설치대수) / 2013년도 이전 설치대수 * 100

```python
cctv_seoul['최근증가율'] = (cctv_seoul['2014년'] +
                          cctv_seoul['2015년'] +
                          cctv_seoul['2016년']) / cctv_seoul['2013년도 이전'] * 100
cctv_seoul.head()
```

```python
# 최근 3개년동안 cctv 설치 증가율이 높은 3개 구
cctv_seoul.sort_values(by='최근증가율', ascending=False).head(3)
```

### 인구데이터의 가공필드 생성 / 기본분석

```python
# 어느 구의 인구가 가장 많은가?
pop_Seoul.head()
pop_Seoul.sort_values(by='인구수', ascending=False).head()

# => 송파구, 강서구, 강남구
```

```python
# 고령자 수가 많은 구
pop_Seoul.sort_values(by='고령자', ascending=False).head()
# 강서구/송파/은평

# 외국인이 많은 구
pop_Seoul.sort_values(by='외국인', ascending=False).head()
# 영등포/구로/금천
```

```python
# 지자체별 전체 인구수 대비 고령자/외국인 비율 계산
pop_Seoul['외국인비율'] = pop_Seoul['외국인']/pop_Seoul['인구수']*100
pop_Seoul['고령자비율'] = pop_Seoul['고령자']/pop_Seoul['인구수']*100
```

```python
pop_Seoul.head()
```

```python
# 인구수 대비 외국인/고령자 비율이 높은 자치구 확인

pop_Seoul.sort_values(by='외국인비율', ascending=False).head()
# 전통적으로 외국인이 많이 거주하는 구 영등포/구로/금천 으로 높게 나타났고
# 중구/용산구는 오피스가 많은 지역이므로 인구는 많지 않지만 파견된 외국인들이 많이 거주하는 지역으로 알려짐


pop_Seoul.sort_values(by='고령자비율', ascending=False).head()
# 강북구는 2017년 당시, 개발이 덜 된 구도심 지역 - 토박이분들이 많은 지역
# 종로, 중구, 용산 대표적인 오피스 지역이고 구도심 - 토박이분들이 많은 지역

# 단, 2017년도 기준의 분석이기 떄문에 현 시점 2023년도에는 많은 변화가 있음
```

### 인구대비 cctv대수의 적정성 확인
- 두 df를 병합하여 새로운 data 생성 후 분석

```python
# 구별 컬럼명을 기준으로 병합
data_result = pd.merge(cctv_seoul, pop_Seoul, on='구별')
data_result.head()
```

```python
# 불필요한 컬럼 삭제
del data_result['2013년도 이전']
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']
```

```python
data_result.head()
```

```python
# df에서 행을 유일하게 구별하는 data(구별)를 행인덱스로 설정
# 시각화 시 효율적
data_result.set_index('구별',inplace=True)
data_result.head()
```

###  인구 관련 data와 cctv 관련 data들 간의 상관성 확인
- 상관성 있는 컬럼이 있는지 확인
    - 서로 관련이 있는 컬럼에 대해 그래프를 그려 적정 cctv 대수 확인
- 상관계수 분석
    - 계수 값이 절대값으로 0.3을 넘어가면 상관관계가 있다.
        - 0.1 이하면 무시
        - 0.7 이상이면 아주 강한 상관관계
        - 0.1 < 계수 <= 0.3 약한 상관관계
    - 값이 -면 음의 상관 : 두 변수의 값 중 하나가 증가면 하나는 감소한다.
    - 값이 + 면 양의 상관 : 두 변수의 값 중 하나가 증가하거나 감소하면 나머지 변수의 값도 증가하거나 감소한다.

```python
data_result.corr()
```

```python
np.corrcoef(data_result['소계'], data_result['인구수'])
# 약한 양의 상관
```

```python
np.corrcoef(data_result['소계'],data_result['고령자비율'])
# 약한 음의 상관
```

```python
np.corrcoef(data_result['소계'], data_result['외국인비율'])
# 관계 없음
```

- cctv 절대수와 전체 인구수가 약하지만 상관관계가 있으므로 두 데이터를 활용해서 인구당 적정 cctv 대수로 추정
    - 고령자 비율도 음의 상관이 있지만 전체 인구를 얘기하지 않으므로 제외

### cctv vs 인구 현황 시각화

```python
data_result.head()
```

```python
# 전체 자치구별 cctv 대수의 차이 확인 시각화
plt.figure(figsize=(10,10))
data_result[['소계']].plot(kind='barh',grid=True)
plt.show()
```

#### 강남구와 도봉구의 cctv 절대 대수의 차이는 굉장히 크다
- 인구수 대비 cctv 설치 비율을 확인해서 절대대수와의 차이를 확인
- 인구수와 설치비율 데이터의 관계 확인
- 가공필드 생성 (인구수대비 cctv 설치비율)

```python
data_result['cctv비율']= data_result['소계']/data_result['인구수']*100
data_result.head()
```

```python
# 자치구별 인구대비 cctv 비율의 차이
data_result['cctv비율'].sort_values().plot(kind='barh',grid=True)
```

- 용산, 중구, 종로구, 강남구는 대표적으로 유동인구가 많은 지역임
    - 거주인원보다 유동인원이 많음
    - 유동인구의 영향으로 인구대비 cctv 비율이 높다

```python
# 인구수 vs cctv 소계의 분산 그래프(scatter)
plt.scatter(data_result.인구수, data_result.소계, s=50)
plt.xlabel('인구수')
plt.ylabel('구별 cctv 수')
plt.grid()
```
