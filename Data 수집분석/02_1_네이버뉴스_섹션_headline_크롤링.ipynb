```python
import pandas as pd
import bs4
import requests
from urllib.request import urlopen
```

```python
# 각 섹션별 링크주소 파일 가져오기
df_menu = pd.read_csv('./crawl_data/naver_news_section.csv',index_col=0)
df_menu.head()
```

### 경제 섹션을 예시로 코드를 작성 후 함수화 진행

```python
url = df_menu['link'][1]  # 경제
url
```

```python
# 경제 섹션 소스 요청
res = urlopen(url)
html = res.read()  # binary로 들어옴
# html
```

```python
bs_obj = bs4.BeautifulSoup(html, 'html.parser')
bs_obj
```

```python
# headline 뉴스 추출
# class : sh_text
head_line = bs_obj.findAll('div', {'class':'sh_text'})
len(head_line)
head_line[0]
```

```python
head_line[0].find('a')  # 기사제목
head_line[0].find('a')['href']  # 헤드라인 뉴스 중 하나의 전체기사 링크
head_line[0].find('a').text  # 헤드라인 뉴스 제목
```

```python
head_line[0].find('div', {'class':"sh_text_lede"}).text  # 뉴스 내용 요약
```

```python
# 전체 헤드라인의 title, link, desc 출력
for head in head_line :
    a_t = head.find('a')
    div_t = head.find('div', {'class':"sh_text_lede"})
    print(a_t.text," : ", a_t['href']," : ", div_t.text)  # 제목, 링크, description
```

```python
# 예외처리
# 전체 헤드라인의 title, link, desc 출력
for head in head_line :
    try :  # 실제 진행되어야 하는 코드
        a_t = head.find('a')
        div_t = head.find('div',{'class':"sh_text_lede"})
        print(a_t.text," : ", a_t['href']," : ", div_t.text)
    except :
        print('에러발생')
```

### 수집한 data를 DataFrame으로 생성

```python
# 빈리스트 생성
headline_title=[]
headline_link=[]
headline_desc=[]
```

```python
for head in head_line :
    try : 
        a_t = head.find('a')
        div_t = head.find('div',{'class':"sh_text_lede"})
        headline_title.append(a_t.text)
        headline_link.append(a_t['href'])
        headline_desc.append(div_t.text)
    except :
        print('에러발생')
```

```python
headline_df = pd.DataFrame({'기사제목':headline_title,
                            '기사링크':headline_link,
                            '기사요약':headline_desc})
headline_df.head()
```
