### 포털사이트 크롤링
- 소스 얻어오기 (네트워크패키지)
- 소스 내에서 특정 문자열(수집하고자하는 data)을 지칭하는 선택자 얻기
    - 크롬개발자도구 사용
- 전체 코드에서 수집하려고 하는 데이터(태그)의 위치를 찾고
    - 태그 파싱 후 데이터 추출

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity="all"
```

```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python
import pandas as pd
import bs4
import requests
from urllib.request import urlopen
```

```python
# 네이트 사이트의 기본 메뉴 텍스트 추출
# 메뉴명, 메뉴 링크
url = 'https://www.nate.com/'

# 네이트 서버에 요청 후 응답 반환
html = urlopen(url)
html

# 소스코드 확인
html_text = html.read()  # HTTPResponse객체에 담겨있는 html코드 반환
html_text  # 바이너리 문자열 반환(한글 변경 후 사용-bs4 객체 사용)
```

```python
# bs4 객체 생성
bs_obj = bs4.BeautifulSoup(html_text, 'html.parser')
bs_obj  # 정적크롤링 가능한 코드가 전달됨 확인
```

```python
# divGnb
# 우클릭 > 검사 : 내가 원하는 태그의 코드 확인
menu = bs_obj.find('div',{'id':'divGnb'})  # copy > copy selector
menu_li = menu.findAll('li')
menu_li
len(menu_li)  # 29개의 li를 찾음
```

```python
# 29개의 li에 대해서 텍스트와 a태그의 href 속성값을 확인
# 원하는 데이터만 있는지 확인
for li in menu_li : 
    print(li.text, li.a['href'])
```

```python
# 29개의 li에 대해서 텍스트와 a태그의 href 속성값을 확인
# 원하는 데이터만 있는지 확인
for li in menu_li : 
    menu_text = li.text
    menu_link = li.a['href']
    if menu_link == 'javascript:;' :  # javascript 제외
        continue
    print(menu_text, menu_link)
```

```python
# 추출된 메뉴에서 중복 제거 후 df로 생성
# 빈 df 생성
nate_menu = pd.DataFrame({'메뉴':[], '링크':[]})
nate_menu
```

```python
index=0  # idx 0부터 시작
for li in menu_li : 
    menu_text = li.text
    menu_link = li.a['href']
    if menu_link == 'javascript:;' or menu_text in nate_menu['메뉴'].values:  # 이미 메뉴에 들어가있다면 패스(중복 제거)
        continue
    temp = pd.DataFrame({'메뉴':menu_text,'링크': menu_link}, index=[index])
    nate_menu = pd.concat([nate_menu, temp])
    index=index+1
```

```python
nate_menu
```

```python
# 수집 데이터 파일로 저장
nate_menu.to_csv('./crawl_data/nate_menu.csv')
```

### 네이버 뉴스 섹션 제목과 링크 크롤링
- 네이버 뉴스는 네이버 정책에 따라 모든 언론사들의 뉴스가 랜덤하게 배치됨
    - 단, 로그인 후 구독을 추가하면 구독한 언론사들의 뉴스가 나옴

```python
from urllib.request import urlopen
import bs4
import pandas as pd
```

```python
url ='https://news.naver.com/'
html = urlopen(url)
bs_obj = bs4.BeautifulSoup(html,'html.parser')
```

```python
bs_obj  # 정적 크롤링이 가능한 페이지
```

```python
# 뉴스 섹션 전체를 포함하는 ul 태그 추출
# li가 포함된 ul 태그 가져오는 방법 주로 사용

# class - Nlnb_menu_list
uls = bs_obj.findAll('ul',{'class':'Nlnb_menu_list'})
len(uls)  # 원소가 1개이므로

ul = bs_obj.find('ul',{'class':'Nlnb_menu_list'})
ul
# type(ul)
```

```python
# 언론사별은 main페이지고 headline기사가 없으므로 제외

ul.findAll('li')[0].find('a') 
```

```python
lis = ul.findAll('li')
# lis[1:]
```

```python
for li in lis[1:] :
    a_tag = li.find('a')
    print(a_tag.text," : ", a_tag['href'])  # href 태그 링크
```

```python
# 위 메뉴 데이터 df로 저장
section=[]
link=[]

# 정치~세계 카테고리는 웹페이지 구조 모두 동일
```

```python
for li in lis[1:] :
    a_tag = li.find('a')
    section.append(a_tag.text)
    link.append(a_tag['href'])
```

```python
len(section)
len(link)
```

```python
col_dict = {'section':section, 'link':link}
news_section_df = pd.DataFrame(col_dict)
news_section_df
```

```python
news_section_df.to_csv('./crawl_data/naver_news_section.csv')
```
