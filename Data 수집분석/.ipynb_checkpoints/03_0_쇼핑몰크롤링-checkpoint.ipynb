### 쇼핑몰 크롤링
- 상품의 시세 파악을 위해서는 쇼핑몰 데이터를 추출하여 파악하는 것이 일반적
- 특정 사이트를 선정하여 매일 상품의 가격 변화를 확인
- 사이트 선정 시 구성이 변하지 않고, 접속량이 일정 수준 이상인 사이트를 선정
- http://jolse.com/category/toners-mists/1019/

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# 1. 소스 추출
url = 'http://jolse.com/category/toners-mists/1019/'
htmls = urlopen(url)
htmls = htmls.read()
```

### SSL CertVerification Error가 발생하면
- 보안연결 시도
- import ssl 패키지를 사용
- ssl 연결 context()를 생성해서 인수로 전달해야 함

```python
import ssl
context = ssl._create_unverified_context()  # 보안체크하지 말 것 (신뢰성 담보)
url = 'http://jolse.com/category/toners-mists/1019/'
htmls = urlopen(url, context=context)
htmls = htmls.read()
```

```python
# 파싱 객체 생성
bs_obj = BeautifulSoup(htmls,'html.parser')
bs_obj
```

#### 화장품 쇼핑몰
- 제품명/가격/세일가격
- 페이징 기능을 사용하는 쇼핑몰이므로 페이지를 다르게 접근해서 모든 제품 크롤링
- 토너/미스트 카테고리에 대해서만 크롤링

```python
# 파싱- 전체 제품을 담고있는 태그 ul 태그를 활용해서 추출
uls = bs_obj.findAll('ul',{'class':'prdList grid5'})
```

```python
len(uls) # best 상품 진열 블럭과 전체제품 진열블럭으로 진열대가 2개
```

```python
# best 상품
# 제품 한개를 담고있는 li 태그 추출
boxes = uls[0].findAll('li')
len(boxes)
```

```python
# 전체 제품의 1 페이지
# 제품 한개를 담고있는 li 태그 추출
boxes = uls[1].findAll('li')
len(boxes)
```

```python
i = 0
title = []
for bx in boxes : 
    try : 
        title.append(bx.find('div',{'class':'description'}).find('strong',{'class':'name'}).text.split(":")[1].strip())
    except : 
        i=i+1
        print('태그없음')
print(i)        
```

```python
len(title)
title
```

```python
# 가격정보
i = 0
price = []
sale_price = []
for bx in boxes : 
    try : 
        div = bx.find('div',{'class':'description'})
        lis = div.find('ul').findAll('li')
        price.append(lis[0].text.split(' ')[2])
        sale_price.append(lis[1].text.split(' ')[2])
    except : 
        i=i+1
#         print('태그없음')
print(i)   
```

```python
len(price), len(sale_price)
```

```python
# 제품명, 원가격, 세일가격정보
i = 0
title = []
price = []
sale_price = []
for bx in boxes : 
    try : 
        div = bx.find('div',{'class':'description'})
        title.append(div.find('strong',{'class':'name'}).text.split(":")[1].strip())
        lis = div.find('ul').findAll('li')
        price.append(lis[0].text.split(' ')[2])
        sale_price.append(lis[1].text.split(' ')[2])
    except : 
        i=i+1
#         print('태그없음')
print(i) 
```

```python
len(title), len(price), len(sale_price)
```

```python
# df 생성
prd_dict = {'품목': title, '가격': price, '세일가격': sale_price}
prd_df = pd.DataFrame(prd_dict, index=range(1, len(title)+1))
prd_df.head()
```
